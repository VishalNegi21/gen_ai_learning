```python
from langchain_groq import ChatGroq
llm = ChatGroq(model="llama-3.3-70b-versatile")
```

### What's the problem

```python
text = "Hello my Name is Vishal Negi , Email is vishalxyz@gmail.com and age is 21"
```

```python
res=llm.invoke(f"Please give me only name , email ,and age from this text :{text}")
```

```python
res.content
```

```python
print(res.content)
```

```python
 type(res.content)  #always string
```

### Structured Output

```python
from pydantic import BaseModel , Field
#define class of structure
class ResponseStructure(BaseModel):
    #properties
    name:str = Field(description="Name ")
    email:str = Field(description="Email id")
    age:int = Field(description="age")
```

```python
structured_llm = llm.with_structured_output(ResponseStructure)
```

```python
res=structured_llm.invoke(text)
res
```

```python
print(res)
```

```python
print(res.age)
print(res.email)
```

```python
#convert to dictionary
res.model_dump()
```

### Multiple levels of Structure

```python
class Movies(BaseModel):
    title:str = Field(description="Movie title")
    year:int = Field(description="Release year")
    cast:str = Field(description="Movie cast")

movie_llm = llm.with_structured_output(Movies)
```

```python
movie_llm.invoke("Give me a hindi action movie")
```

```python
# asked for 3 movies still gives only one
movie_llm.invoke("Give me three hindi action movies")
```

#### WHat if we want list of movies not a single movie

```python
from typing import List
class Movie(BaseModel):
    title:str = Field(description="Movie title")
    year:int = Field(description="Release year")
    cast:str = Field(description="Movie cast")

class MultipleMovies(BaseModel):
    movies:List[Movies]
```

```python
movie_llm=llm.with_structured_output(MultipleMovies)
res=movie_llm.invoke("Give me 3 hindi action movie")
res
```

```python
res.movies
```

```python
res.model_dump()
```
