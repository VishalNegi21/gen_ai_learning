```python
from langchain_groq import ChatGroq
from langchain.agents import create_agent
```

```python
from langchain_community.utilities import GoogleSerperAPIWrapper
```

```python
llm = ChatGroq(model="openai/gpt-oss-20b")
search = GoogleSerperAPIWrapper()
```

```python
search.run("address of MSIT Delhi")
```

### connect search and llm (using agent)

```python
agent = create_agent(
    model=llm,
    tools=[search.run],
    system_prompt="You are a agent who can search the answers for any question on google"
)
```

```python
question = "What is the address , email and phone number of MSIT Delhi"
```

```python
res=agent.invoke({"messages":[{"role":"user","content":question}]})
```

```python
print(res["messages"][-1].content)
```
