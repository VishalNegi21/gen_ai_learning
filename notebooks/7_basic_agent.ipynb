### Create Tool

```python
from langchain.tools import tool
```

```python
@tool
def add_numbers(a:int , b:int):
    #docstring
    """
        it will add the numbers
        Args:
            a:Number 1
            b:Number 2
    """
    return a+b
```

```python
add_numbers.invoke({"a":4,"b":9})
```

```python
@tool
def multiply_numbers(a:int , b:int):
    #docstring
    """
        it will multiply the numbers
        Args:
            a:Number 1
            b:Number 2
    """
    return a*b
```

```python
multiply_numbers.invoke({"a":8,"b":6})
```

### Connect LLM with tool


```python
from langchain_groq import ChatGroq
llm = ChatGroq(model="llama-3.3-70b-versatile")
```

```python
from langchain.agents import create_agent
```

```python
agent = create_agent(
    model=llm,
    tools= [add_numbers,multiply_numbers],
    system_prompt="You are a mathematics teacher , always use tools for calculation"
)
```

```python
response = agent.invoke({"messages":[{"role":"user" , "content":"What is 2+3*2"}]})
```

```python
for res in response["messages"]:
    print(res )
    print("\n")
```
