```python
from langchain_groq import ChatGroq
llm = ChatGroq(model="llama-3.3-70b-versatile")
```

```python
prompts =[{
    "role":"user" ,"content":"Hello My name is Vishal"
}]
```

```python
res = llm.invoke(prompts)
print(res.content)
```

```python
prompts=[
    {"role":"user" , "content":"Hello My name is Vishal"},
    {"role":"ai","content":"Hello Vishal, it's nice to meet you. Is there something I can help you with or would you like to chat?"},
    {"role":"user" , "content":"What is my name?"}
]
```

```python
res = llm.invoke(prompts)
print(res.content)
```

### Manage History

```python
history=[]
while True:
    query = input("User :")
    if query.lower() in ["bye","exit","quit"]:
        print("BYE")
        break

    history.append({"role":"user" , "content":query})
    print("User :",query)

    res = llm.invoke(history)
    history.append({"role":"ai" , "content":res.content})
    print("AI" ,res.content ,"\n")
```
