### Gemini model

```python
from langchain_google_genai import ChatGoogleGenerativeAI
lm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")   
```

```python
response=llm.invoke("What is generative ai , explain in short?")
print(response.content)
```

### GROQ

### Static prompts

```python
from langchain_google_genai import ChatGoogleGenerativeAI
lm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")   
```

```python
prompts=[
    ("system","You are an AI Engineer."),
    ("user" , " how to sort an array?"),
]
```
